```python
import pymysql
import pandas as pd
import os

# 连接数据库
conn = pymysql.connect(
    host='localhost',      # 数据库在自己电脑上，所以是localhost
    user='root',           # 默认用户名
    password='jiaxin0411', # 我的密码
    database='olist',      # 我们刚建的数据库
    charset='utf8mb4'      # 支持中文和特殊字符
)

print("数据库连接成功！")
```

```python
# CSV文件所在文件夹
path = r"C:\Users\jiaxin\Downloads\archive"

# 定义每个CSV文件对应哪张数据库表
files = {
    "olist_customers_dataset.csv": "customers",
    "olist_orders_dataset.csv": "orders",
    "olist_order_items_dataset.csv": "order_items",
    "olist_order_payments_dataset.csv": "order_payments",
    "olist_order_reviews_dataset.csv": "order_reviews",
    "olist_products_dataset.csv": "products",
    "olist_sellers_dataset.csv": "sellers",
    "olist_geolocation_dataset.csv": "geolocation"
}

# 循环把每个CSV导入对应的表
for filename, tablename in files.items():
    filepath = os.path.join(path, filename)
    df = pd.read_csv(filepath)
    df.to_sql(tablename, con=conn, if_exists='append', index=False)
    print(f"{tablename} 导入成功，共 {len(df)} 行")

print("\n全部导入完成！")
```
