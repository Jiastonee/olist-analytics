```python
import pandas as pd
from sqlalchemy import create_engine

# 连接数据库
engine = create_engine('mysql+pymysql://root:jiaxin0411@localhost/olist?charset=utf8mb4')

# 把主要的表读进来
orders = pd.read_sql('SELECT * FROM orders', engine)
customers = pd.read_sql('SELECT * FROM customers', engine)
order_items = pd.read_sql('SELECT * FROM order_items', engine)
order_payments = pd.read_sql('SELECT * FROM order_payments', engine)

print("数据读取成功！")
print(f"orders: {orders.shape}")
print(f"customers: {customers.shape}")
print(f"order_items: {order_items.shape}")
print(f"order_payments: {order_payments.shape}")
```

```python
# 检查每张表的缺失值
print("=== orders 缺失值 ===")
print(orders.isnull().sum())

print("\n=== customers 缺失值 ===")
print(customers.isnull().sum())

print("\n=== order_items 缺失值 ===")
print(order_items.isnull().sum())

print("\n=== order_payments 缺失值 ===")
print(order_payments.isnull().sum())
```

```python
# 检查重复值
print("=== 重复值检查 ===")
print(f"orders 重复行: {orders.duplicated().sum()}")
print(f"customers 重复行: {customers.duplicated().sum()}")
print(f"order_items 重复行: {order_items.duplicated().sum()}")
print(f"order_payments 重复行: {order_payments.duplicated().sum()}")

# 检查价格异常值
print("\n=== order_items 价格统计 ===")
print(order_items['price'].describe())
```
